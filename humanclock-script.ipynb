```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
import os
import cssutils
from PIL import Image
from io import StringIO,BytesIO
import datetime
import requests
import time
```

```python
url_humanclock = 'https://www.humanclock.com/'
```

```python
driver = webdriver.Firefox()
driver.get(url_humanclock)
driver.implicitly_wait(10)
# import timeit
# start_time = timeit.default_timer()
# code you want to evaluate
starttime=time.time()
while True:
    driver.refresh()
    soup = BeautifulSoup(driver.page_source, 'lxml')
    now = datetime.datetime.now()
    print(f'Time Scrapped {now.hour}:{now.minute}')
    imgs = soup.find_all(class_= "spriteImage")
    url_images = set()
    image_number = 0

    for img in imgs:
        div_style = cssutils.parseStyle(img['style'])
        url = div_style['background-image']
        url = url.replace('url(', '').replace(')', '')
        if len(url) > 1:
            image_number+=1
            url_images.add(url)

    url_images = list(url_images)
    download_picture(url_images[0])
    print(f'Image saved {url_images[0]}')
    time.sleep(55)    
    #     break
# elapsed = timeit.default_timer() - start_time
# elapsed
# driver.quit()
```

```python
def download_picture(url_picture):
    response = requests.get(url_picture, stream=True)

    image_path = get_picture_path()
    if os.path.exists(image_path):
        print(f'Image {image_path} already exists')
        return

    image = Image.open(BytesIO(response.content))

    image.save(image_path)
```

```python
def get_picture_path():
    now = datetime.datetime.now()
    image_name = f'{now.hour}:{now.minute}.jpg'
    str_hour = str(now.hour).rjust(2,'0')
    str_minute = str(now.minute).rjust(2,'0')
    path = f'dataset/{str_hour}/{str_minute}/{image_name}'
    return path
```

```python
def create_directories_structure(base_dir='dataset/'):
    if not os.path.exists(base_dir):
        os.mkdir(base_dir)
        
    for hour in range(1,25):
        str_hour = '00' if hour == 24 else str(hour).rjust(2, '0')
        if not os.path.exists(base_dir+str_hour):
            os.mkdir(base_dir+str_hour)
            print(f'Directory {str_hour}/ created')
        else:
            print(f'Directory {str_hour}/ already exists')
        
        for minute in range(0, 60):
            str_minute = str(minute).rjust(2, '0')
            path_minute = f'{base_dir}{str_hour}/{str_minute}'
            if not os.path.exists(path_minute):
                os.mkdir(path_minute)
                print(f'Directory {path_minute}/ created')
            else:
                print(f'Directory {path_minute}/ already exists')
```
